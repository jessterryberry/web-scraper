```python
import requests
from bs4 import BeautifulSoup

URL = "https://niagaraconstruction.org/members/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find(class_="staffContent")

company_elements = results.find_all("a")

for company_element in company_elements:
    web_link = company_element.get("href")
    if "alphaSet" in web_link : continue

    member_page = requests.get("https://niagaraconstruction.org/" + web_link)
    member_soup = BeautifulSoup(member_page.content, "html.parser")

    info_fields = member_soup.find_all("div", class_="staffTitleDiv")
    member_props = ["" for i in range(12)]
    i = 0

    for info_field in info_fields:
        text = (info_field.text.split(":", 1)[1].strip().replace('\n', '').replace('  ', ''))
        header = (info_field.text.split(":", 1)[0].strip().replace('\n', '').replace('  ', '')) 
        if header == "Name" : member_props[0] = text
        if header == "Address" : member_props[1] = text
        if header == "City" : member_props[2] = text
        if header == "Province" : member_props[3] = text
        if header == "Postal Code" : member_props[4] = text
        if header == "Country" : member_props[5] = text
        if header == "Telephone" : member_props[6] = text
        if header == "Ext" : member_props[7] = text
        if header == "Fax" : member_props[8] = text
        if header == "Website" : member_props[9] = text
        if header == "General Email" : member_props[10] = text
        if header == "CATEGORIES" : member_props[11] = text
        i += 1
        if i == 12 : break
    print(";".join(member_props))    
```
